```python
def random_dates(start, end, out_format='datetime', n=100):
    (divide_by, unit)= (10**9, 's') if out_format=='datetime' else (24*60*60*10**9, 'D')
    start_u = start.value//divide_by
    end_u = end.value//divide_by
    
    return pd.to_datetime(np.random.randint(start_u, end_u, n), unit=unit)
```

```python
import pandas as pd
start_date = pd.to_datetime('2000-01-01')
end_date = pd.to_datetime('2015-01-01')

random_dates(start_date, end_date, out_format='datetime', n=100)
```

```python
import numpy as np
import pandas as pd
from sklearn.base import BaseEstimator, TransformerMixin

class retainFields(BaseEstimator, TransformerMixin):
    def __init__(self, variables):
        self.variables = variables
        
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        X = X[self.variables]
        return X
    
        
class handleNulls(BaseEstimator, TransformerMixin):
    def __init__(self, variables=None):
        self.variables= variables
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        X = X.copy()
        X = X.fillna(method='ffill')
        X = X.fillna(method='bfill')
#         X[self.variables[0]] = X[self.variables[0]].fillna(method='ffill')
#         X[self.variables[1]] = X[self.variables[1]].fillna(method='ffill')
#         X[self.variables[2]] = X[self.variables[2]].fillna(method='ffill')
        return X

class fillOdom(BaseEstimator, TransformerMixin):
    def __init__(self, variable=None):
        self.variable = variable
    
    def fit(self, X,y=None):
        self.odom_mode = X[self.variable].mode()[0]
        return self
    
    def transform(self, X):
        X=X.copy()
        
        X[self.variable] = X[self.variable].fillna(self.odom_mode)
        return X
    
class Manu_Encoder(BaseEstimator, TransformerMixin):
    def __init__(self, variables=None):
        self.variables= variables
        
    def fit(self, X,y=None):
        self.man_dict = {}
        self.a = X[self.variables].unique()
        for i in range(len(self.a)):
            self.man_dict[self.a[i]]=i 
               
        return self
    
    def transform(self, X):
        X = X.copy()
        
        X[self.variables] = X[self.variables].replace(self.man_dict)     
        return X

class Trans_Encoder(BaseEstimator, TransformerMixin):
    def __init__(self, variables=None):
        self.variables= variables
        
    def fit(self, X,y=None):
        self.trans_dict = {}
        self.a = X[self.variables].unique()
        for i in range(len(self.a)):
            self.trans_dict[self.a[i]]=i 
               
        return self
    
    def transform(self, X):
        X = X.copy()
        
        X[self.variables] = X[self.variables].replace(self.trans_dict)     
        return X

class Type_Encoder(BaseEstimator, TransformerMixin):
    def __init__(self, variables=None):
        self.variables= variables
        
    def fit(self, X,y=None):
        self.type_dict = {}
        self.a = X[self.variables].unique()
        for i in range(len(self.a)):
            self.type_dict[self.a[i]]=i 
              
        return self
    
    def transform(self, X):
        X = X.copy()
         
        X[self.variables] = X[self.variables].replace(self.type_dict)     
        return X
    
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

# from regression_model.processing import preprocessors as pp
# from regression_model.config import config

# import logging


# _logger = logging.getLogger(__name__)
cat_variables = ['manufacturer', 'transmission', 'type'] 

# ext_vars = ['price', 'year']
null_vars = ['manufacturer', 'type', 'transmission']

vars_to_retain = ['manufacturer', 'year','odometer', 'transmission', 'type', 'ori_cost_prc']

# variables_to_drop = ['url','image_url', 'region_url', 'lat', 'long', 'region', 'description', 'vin', 'county', 'model']


price_pipe = Pipeline([
        ("retainFields", retainFields(vars_to_retain)),
        ("fillOdom", fillOdom('odometer')),
        ("handleNulls", handleNulls()),
        ("Manu_Encoder", Manu_Encoder('manufacturer')),
        ("Type_Encoder", Type_Encoder('type')),
        ("Trans_Encoder", Trans_Encoder('transmission')),
        ("scaler", StandardScaler()),
        ("Linear_model", LogisticRegression()),
    ])
```
