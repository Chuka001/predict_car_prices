```python
import numpy as np
import pandas as pd
from sklearn.base import BaseEstimator, TransformerMixin

class DropUnecessaryFeatures(BaseEstimator, TransformerMixin):
    def __init__(self, variables=None):
        self.variables = variables
    
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        # encode labels
        X = X.copy()
        X = X.drop(self.variables, axis=1)

        return X
    
class retainFields(BaseEstimator, TransformerMixin):
    def __init__(self, variables):
        self.variables = variables
        
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        X = X[self.variables]
        return X
    
        
class handleNulls(BaseEstimator, TransformerMixin):
    def __init__(self, variables=None):
        self.variables= variables
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        X = X.copy()
        X = X.fillna(method='ffill')
        return X

class fillOdom(BaseEstimator, TransformerMixin):
    def __init__(self, variable='odometer'):
        self.variable = variable
    
    def fit(self, X,y=None):
        self.odom_mode = X[self.variable].mode()[0]
        return self
    
    def transform(self, X):
        X=X.copy()
        X[self.variable] = X[self.variable].fillna(self.odom_mode)
        return X

class Manu_Encoder(BaseEstimator, TransformerMixin):
    def __init__(self, variables=None):
        self.variables= variables
        
    def fit(self, X,y=None):
               
        return self
    
    def transform(self, X):
        X = X.copy()
        man_dict = {}
        a = X[self.variables].unique()
        for i in range(len(a)):
            man_dict[[i]]=i 
        X[self.variables] = X[self.variables].replace(man_dict)     
        return X

class Trans_Encoder(BaseEstimator, TransformerMixin):
    def __init__(self, variables=None):
        self.variables= variables
        
    def fit(self, X,y=None):
               
        return self
    
    def transform(self, X):
        X = X.copy()
        trans_dict = {}
        a = X[self.variables].unique()
        for i in range(len(a)):
            trans_dict[a[i]]=i 
        X[self.variables] = X[self.variables].replace(trans_dict)     
        return X

class Type_Encoder(BaseEstimator, TransformerMixin):
    def __init__(self, variables=None):
        self.variables= variables
        
    def fit(self, X,y=None):
              
        return self
    
    def transform(self, X):
        X = X.copy()
        type_dict = {}
        a = X[self.variables].unique()
        for i in range(len(a)):
            type_dict[a[i]]=i  
        X[self.variables] = X[self.variables].replace(type_dict)     
        return X
    
# class CategoricalEncoder(BaseEstimator, TransformerMixin):
#     def __init__(self, variables=None):
#         self.variables= variables
        
#     def fit(self, X,y=None):
#         self.man_dict = man_dict
#         self.type_dict = type_dict
#         self.trans_dict = trans_dict
#         return self
    
#     def transform(self, X):
#         X = X.copy()
#         X[self.variables[0]] = X[self.variables[0]].replace(self.man_dict)
#         X[self.variables[1]] = X[self.variables[1]].replace(self.trans_dict)
#         X[self.variables[2]] = X[self.variables[2]].replace(self.type_dict)
        
#         return X
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

# from regression_model.processing import preprocessors as pp
# from regression_model.config import config

# import logging


# _logger = logging.getLogger(__name__)
cat_variables = ['manufacturer', 'transmission', 'type'] 

# ext_vars = ['price', 'year']
# null_vars = ['manufacturer', 'type', 'odometer', 'transmission']

vars_to_retain = ['manufacturer', 'year','odometer', 'transmission', 'type', 'ori_cost_prc']

variables_to_drop = ['url','image_url', 'region_url', 'lat', 'long', 'region', 'description', 'vin', 'county', 'model']


price_pipe = Pipeline(
    [
        ("DropUnecessaryFeatures", DropUnecessaryFeatures(variables_to_drop )),
        ("retainFields", retainFields(vars_to_retain)),
        ("fillOdom", fillOdom()),
        ("handleNulls", handleNulls()),
        ("Manu_Encoder", Manu_Encoder('manufacturer')),
        ("Type_Encoder", Type_Encoder('type')),
        ("Trans_Encoder", Trans_Encoder('transmission')),
        ("scaler", StandardScaler()),
        ("Linear_model", LogisticRegression()),
    ]
)
```
